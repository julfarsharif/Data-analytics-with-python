```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
normal_df=pd.DataFrame({'x':np.random.randint(-10,10,1000),'y':np.random.randint(-10,10,1000),'z':np.random.randint(-5,5,1000)})
```

```python
normal_df
```

```python
ax=plt.axes(projection='3d')

ax.scatter3D(normal_df.x,normal_df.y,normal_df.z)

plt.figure()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normal=Normalizer()
```

```python
normal_df=normal.fit_transform(normal_df)
```

```python
normal_df
```

```python
norm_df=pd.DataFrame(normal_df,columns=['x1','x2','x3'])
```

```python
norm_df
```

```python
ax=plt.axes(projection='3d')

ax.scatter3D(norm_df.x1,norm_df.x2,norm_df.x3)

plt.figure()
```

```python
datas={'First score':[100,30,np.nan,74],'Second score':[90,np.nan,76,98],'Third score':[np.nan,67,98,87]}
```

```python
scores=pd.DataFrame(datas)
```

```python
scores
```

```python
scores.isnull()
```

```python
scores.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
i=SimpleImputer(missing_values=np.nan,strategy='median')
```

```python
i.fit(scores)
```

```python
i.transform(scores)
```

```python
i=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
i.fit(scores)
```

```python
i.transform(scores)
```
