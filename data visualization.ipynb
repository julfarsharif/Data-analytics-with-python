# creating data

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = ['Legend','Beast','End game','PS 1','Valimai','Pathan','Sita ramam']
percentage = [9,30,95,80,30,90,99]
```

```python
plt.scatter(movies,percentage)
```

```python
plt.bar(movies,percentage)
```

```python
plt.plot(movies,percentage,linestyle='solid')
```

```python
plt.plot(movies,percentage,linestyle='dashdot')
```

```python
plt.pie(percentage,labels=movies)
```
